```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris['species'].value_counts()
```

```python
iris.head()
```

```python
iris.describe()
```

```python
col = 'sepal_length'
sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variável {col}')
```

```python
for col in iris.drop(columns='species'):
  sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição de variável {col}')
  plt.show()
```

```python
for col in iris.drop(columns='species'):
  sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f'Distribuição de variável {col}')
  plt.show()
```

```python
for col in iris.drop(columns='species'):
  sns.barplot(data=iris, x=iris['species'], hue=iris['species'], y = col, ci = 90)
  plt.show()
```
